```python
import numpy as np
import matplotlib.pyplot as plt

# metrics: sup_f1, sup_em, ans_f1, ans_em
_2hop = np.array([85.5, 70.8, 58.1, 48.0])
_3hop1 = np.array([83.6, 50.9, 52.2, 42.3])
_3hop2 = np.array([76.6, 33.3, 30.3, 24.5])
_4hop1 = np.array([76.2, 36.2, 45.5, 42.7])
_4hop2 = np.array([68.6, 4.7, 12.9, 9.4])
_4hop3 = np.array([78.1, 25.3, 24.8, 18.9])

x = np.arange(len(_2hop))
metrics = ["Sup-F1", "Sup-EM", "Ans-F1", "Ans-EM"]

# bar
plt.figure(figsize=(16,6), dpi=80)
width = 0.10
gap = 0.14
plt.bar(x-2.5*gap, _2hop, width=width, label="2hop", color="#EBC999")
plt.bar(x-1.5*gap, _3hop1, width=width, label="3hop1", color="#8CA9D3")
plt.bar(x-0.5*gap, _3hop2, width=width, label="3hop2", color="#F26659")
plt.bar(x+0.5*gap, _4hop1, width=width, label="4hop1", color="#EBC999")
plt.bar(x+1.5*gap, _4hop2, width=width, label="4hop2", color="#8CA9D3")
plt.bar(x+2.5*gap, _4hop3, width=width, label="4hop3", color="#F26659")

# text
for i, j in zip(x, _2hop):
    plt.text(i-0.40, j+1, _2hop[i])
for i, j in zip(x, _3hop1):
    plt.text(i-0.26, j+1, _3hop1[i])
for i, j in zip(x, _3hop2):
    plt.text(i-0.12, j+1, _3hop2[i])
for i, j in zip(x, _4hop1):
    plt.text(i+0.02, j+1, _4hop1[i])
for i, j in zip(x, _4hop2):
    plt.text(i+0.16, j+1, _4hop2[i])
for i, j in zip(x, _4hop3):
    plt.text(i+0.30, j+1, _4hop3[i])
# for i, j in zip(x, bart_hop3):
#     plt.text(i-0.11, j+0.01, bart_hop3[i])
# for i, j in zip(x, bart_hop4):
#     plt.text(i+0.16, j+0.01, bart_hop4[i])

plt.xticks(x, metrics)
plt.legend()

plt.show()
```
