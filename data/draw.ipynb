```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4), dpi=100)

# Doc
train = np.load("doc_len_train.npy")
dev = np.load("doc_len.npy")

plt.subplot(121)
n1, bins1, _ = plt.hist(train, color="#8CA9D3", label="train", bins=10, range=(0, 500), rwidth=0.9, edgecolor="#888")
n2, bins2, _ = plt.hist(dev, color="#F26659", label="dev", bins=10, range=(0, 500), rwidth=0.9, edgecolor="#222", alpha=0.9)

for i in range(len(n1)):
    plt.text(bins1[i]+5, n1[i]+250, int(n1[i]))
    if i < 6:
        plt.text(bins2[i]+5, n2[i]+100, int(n2[i]))

plt.xlabel("doc_len")
plt.legend()

# Query
train = np.load("query_len_train.npy")
dev = np.load("query_len.npy")

plt.subplot(122)
n1, bins1, _ = plt.hist(train, color="#8CA9D3", label="train", bins=5, range=(0, 25), rwidth=0.9, edgecolor="#888")
n2, bins2, _ = plt.hist(dev, color="#F26659", label="dev", bins=5, range=(0, 25), rwidth=0.9, edgecolor="#222", alpha=0.9)

for i in range(len(n1)):
    plt.text(bins1[i]+1.5, n1[i]+300, int(n1[i]))
    if i in (1,2,3):
        plt.text(bins2[i]+1.5, n2[i]+150, int(n2[i]))

plt.xlabel("query_len")
plt.legend()

plt.show()
```
